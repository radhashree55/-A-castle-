```python
import csv
import pandas as pd
import statistics as st
import plotly.express as px

df = pd.read_csv("savings_data.csv")
fig = px.scatter(df, y="quant_saved",color = "highschool_completed")
fig.show()
```

```python
with open("savings_data.csv",newline = "") as f:
  reader = csv.reader(f)
  savingsData = list(reader)
savingsData.pop(0)
totalEntries = len(savingsData)
totalHighSchoolers = 0
for data in savingsData:
  if int(data[2])==1:
    totalHighSchoolers+=1
totalNonHighSchoolers = (len(savingsData)-totalHighSchoolers)

import plotly.graph_objects as go
fig = go.Figure(go.Bar(x=["High Schoolers","Non-High-Schoolers"],y = [totalHighSchoolers, totalNonHighSchoolers]))
fig.show()
```

```python
allSavings = []
for data in savingsData:
  allSavings.append(float(data[0]))

import statistics as st
mean = st.mean(allSavings)
median = st.median(allSavings)
mode = st.mode(allSavings)
print("Mean of the data is",mean)
print("Median of the data is",median)
print("Mode of the data is",mode)
```

```python
highSchoolers=[]
nonHighSchoolers=[]
for data in savingsData:
  if int(data[3])==1:
    highSchoolers.append(float(data[0]))
  else:
    nonHighSchoolers.append(float(data[0]))

mean1 = st.mean(highSchoolers)
median1 = st.median(highSchoolers)
mode1 = st.mode(highSchoolers)
print("Data for people who are Highschoolers")
print("Mean of the data is",mean1)
print("Median of the data is",median1)
print("Mode of the data is",mode1)

print("Data for people who are not Highschoolers")
mean2 = st.mean(nonHighSchoolers)
median2 = st.median(nonHighSchoolers)
mode2 = st.mode(nonHighSchoolers)
print("Mean of the data is",mean2)
print("Median of the data is",median2)
print("Mode of the data is",mode2)

stdev = st.stdev(allSavings)
stdev1= st.stdev(highSchoolers)
stdev2=st.stdev(nonHighSchoolers)
print("Standard Deviation of the Data is",stdev)
print("Standard Deviation of the Data for people who are highschoolers is",stdev1)
print("Standard Deviation of the Data for people who are non highschoolers is",stdev2)
```
